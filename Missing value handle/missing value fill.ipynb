***Null value filling***

```python
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.fillna(10)
```

```python
df.info()
```

**backword filling**

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='bfill',axis=1)
```

**Forword filling**

```python
df.fillna(method='ffill')
```

```python
df.fillna(method = 'ffill',axis=1)
```

**filling with mode**

```python
#df['Age'].mode()[0]
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
df.tail()
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df.tail()
```

```python
df.isnull().sum()
```

all colum filling with mode in categorical

```python
data_set = pd.read_csv('test.csv')
```

```python
data_set.isnull().sum()
```

```python
data_set.info()
```

```python
for i in data_set.select_dtypes(include='object').columns:
    data_set[i].fillna(data_set[i].mode()[0],inplace=True)
data_set
```

```python
data_set.isnull().sum()
```

****filling missing value with scikit-learn***

```python
from sklearn.impute import SimpleImputer
```

```python
dataset = pd.read_csv('test.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.select_dtypes(include='float64').columns
```

```python
si = SimpleImputer(strategy='mean')
si.fit_transform(dataset[['Age', 'Fare']])
```

```python
ar=si.fit_transform(dataset[['Age', 'Fare']])
```

```python
new=pd.DataFrame(ar,columns=dataset.select_dtypes(include='float64').columns)
new.tail()
```

```python
new.isnull().sum()
```
