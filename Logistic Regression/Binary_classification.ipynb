```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

* now we are drop EstimatedSalary this column becouse we are work age and purchased colimn

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.drop(columns=['EstimatedSalary'],inplace=True)
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age',y="Purchased",data=df)
```

* This graph we can see its follow logistic regression

```python
x= df[['Age']]
y= df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

check

```python
lr.predict([[35]])
```

* now check how of this line in this dataset

```python
sns.scatterplot(x='Age',y="Purchased",data=df)
sns.lineplot(x='Age',y=lr.predict(x),data = df,c='red')
```
