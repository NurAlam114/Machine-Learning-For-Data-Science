# Binary_classification_polynomial_input

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("polynomial_classification_data.csv")
df.head()
```

```python
sns.scatterplot(x='data1',y = 'data2',data=df,hue='output')
```

```python
x=df.iloc[:,:-1]
y=df['output']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

* here we get very low accuracy

```python
from mlxtend.plotting import plot_decision_regions 
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = lr)
```

* Here we can see in this linearly Separable here some mistake in this graph and here miss classification.

* if data is not linearly sparable then we are doing polynomial regression
