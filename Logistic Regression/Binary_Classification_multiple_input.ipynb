# Binary_Classification_multiple_input

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads_multiple_input.csv')
df.head()
```

* now we check its lniearly Separable or not if its linearly separable then we start work. 

```python
sns.scatterplot(x='Age',y='EstimatedSalary',data=df,hue='Purchased')
plt.legend(loc=1)
```

```python
x= df.iloc[:,:-1]
y= df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

* here 100% Accuracy result provide.thats a very good result  

now we see decision boundary,here we can see how to create Separable line.

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
```
