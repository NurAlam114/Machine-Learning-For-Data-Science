# Feature Scaling --- Normalization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('scaling_dataset.csv')
dataset
```

Then check any null value or not .if null value are exist then handle 
this null value with mean.

```python
dataset.describe()
```

```python
sns.distplot(dataset['Savings (USD)'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit_transform(dataset[["Savings (USD)"]])
```

```python
dataset ['Savings (USD)_ms'] = ms.fit_transform(dataset[["Savings (USD)"]])
dataset
```

```python
plt.subplot(1,2,1)
plt.title('befor')
sns.distplot(dataset['Savings (USD)'])

plt.subplot(1,2,2)
plt.title('after')
sns.distplot(dataset['Savings (USD)_ms'])
```
