```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('outlier_test_dataset.csv')
df.head()
```

```python
df.shape
```

```python
sns.boxplot(x="value", data=df)
plt.title("Boxplot of Feature2 (Before Outlier Removal)")
plt.show()
```

*we remove the outlier with formula  

Z_score = X - mean  /  std

```python
z_scores = (df['value'] - df['value'].mean()) / df['value'].std()
```

```python
df_clean = df[np.abs(z_scores) < 3]
```

```python
sns.boxplot(x="value", data=df_clean)
plt.title("Boxplot of Feature2 (After Outlier Removal)")
plt.show()
```
