```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import numpy as np
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='cgpa',y="placement_exam_marks",data = df,hue= 'placed')
```

```python
x = df.iloc[:,:-1]
y = df['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel='linear')
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
```

* we will use  kernel='linear' when data are separable. our data is not separable so we are try another all karnel and check

```python
sv1 = SVC(kernel="sigmoid")
sv1.fit(x_train,y_train)
```

```python
sv1.score(x_test,y_test)*100
```

```python
sv1.score(x_train,y_train)*100
```
