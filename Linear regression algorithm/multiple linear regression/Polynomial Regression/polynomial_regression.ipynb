# ***Polynomial Regression** 

*polynomial regression dont follow linearity 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('polynomial_regression_sample.csv')
df.head()
```

```python
plt.scatter(df['x'],df['y'])
```

or

```python
sns.scatterplot(x='x',y='y',data = df)
```

```python
df.corr()
```

```python
x = df[['x']]
y = df['y']
```

* now convert linear regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)

x = pf.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

* now for polynomial

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

* here we can see 99% train accuricy.

```python
predict = lr.predict(x)
```

```python
plt.scatter(df['x'],df['y'])
plt.plot(df['x'],predict,c="red")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Orginal','predict'])
```

* we can see this red line this is very good train up in this dataset. 

* now we can check.

```python
test = pf.fit_transform([[-10]])
test
```
