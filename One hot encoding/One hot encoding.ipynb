****one hot encoding***

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset = pd.read_csv('test.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
en_data =dataset[['Sex']]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
ohe=OneHotEncoder()
ohe.fit_transform(en_data).toarray()
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=['Sex_female','Sex_male'])
```

```python
ohe = OneHotEncoder(drop='first')
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=['Sex_male'])
```
